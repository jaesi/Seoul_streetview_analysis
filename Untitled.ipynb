```python
from tqdm import tqdm

# ...

# 각 모델의 성능을 저장할 리스트 초기화
model_names = []
cross_val_scores = []
r2_scores = []

# 각 모델에 대한 5-fold cross-validation 점수와 R-squared 점수 계산
for model_name, model in tqdm(models.items()):  # tqdm으로 래핑하여 진행 상황 표시
    # 5-fold cross-validation 점수 계산
    scores = cross_val_score(model, X_train, y_train, cv=5, scoring='r2')
    mean_cv_score = np.mean(scores)
    
    # 모델을 트레이닝 데이터로 학습
    model.fit(X_train, y_train)
    
    # 테스트 데이터에 대한 예측 및 R-squared 점수 계산
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    
    # 결과를 리스트에 저장
    model_names.append(model_name)
    cross_val_scores.append(mean_cv_score)
    r2_scores.append(r2)

# ...
```

```python
# 시각화: 각 모델의 5-fold Cross-Validation 평균 점수와 R-squared 점수를 막대 그래프로 표시
plt.figure(figsize=(12, 6))
plt.bar(model_names, cross_val_scores, label='5-fold Cross-Validation avg score', color='skyblue')
plt.bar(model_names, r2_scores, label='R-square score', color='lightcoral', alpha=0.7)
plt.xlabel('Model')
plt.ylabel('Score')
plt.title('5-fold Cross-Validation Average Scores와 R-squared Scores')
plt.legend(loc='lower right')
plt.ylim(0, 1)  # 점수 범위를 0에서 1로 설정 (R-squared의 범위)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
